```python
plt.figure(figsize = (8,6))
sns.histplot(df['charges'] , kde = True , bins = 20)
plt.grid(True)
plt.title("distribution of charges column")
plt.show()
```

```python
cols = ['charges' , 'age' , 'bmi']
fig , axes = plt.subplots(1 , 3 , figsize = (15 , 4))

for ax , col in zip(axes , cols):
    sns.histplot(df[col],kde = True , ax = ax)
    ax.set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data = df , x = 'age' , y = 'charges' , hue = 'age')

plt.title("age vs charges scatter plot")
plt.show()
```

```python
sns.scatterplot(data = df['age'] )

plt.title("age")
plt.show()
```

```python
sns.scatterplot(data = df , x = 'bmi' , y = 'charges' , hue = 'bmi')

plt.title("bmi vs charges scatter plot")
plt.show()
```

```python
sns.boxplot(data=df[['smoker', 'charges']])
plt.title('smoker vs charges boxplot')
plt.show()
```

```python
corr = df[['age','bmi' , 'charges' , 'children']].corr()
sns.heatmap(corr)
plt.title("correaltion heatmap")
plt.show()
```
